```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
import pandas as pd
import time

# 웹드라이버 생성
driver = webdriver.Chrome()

place_list = []
for page in range(1,48):
    time.sleep(3)
    # 웹페이지 접근
    url = f"https://korean.visitseoul.net/attractions?curPage={page}&srchType=&srchOptnCode=&srchCtgry=&sortOrder=2&srchWord=&radioOptionLike=TURSM_AREA_8"
    driver.get(url)
    # HTML 문자열을 파싱하여 BeautifulSoup 객체 생성
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    

    # 모든 아이탬 리스트 생성하기
    items = soup.find_all('li', class_='item')

    for item in items:
        time.sleep(1)
        link = item.find('a')['href']
        full_link = "https://korean.visitseoul.net" + link
        # 새로운 탭에서 링크 열기
        driver.execute_script("window.open(arguments[0]);", full_link)
        driver.switch_to.window(driver.window_handles[1])

        # 새로운 탭 데이터 추출
        html_detail = driver.page_source
        soup_detail = BeautifulSoup(html_detail, 'html.parser')


        # 기존 페이지 데이터 추출
        place_name = item.find('span', class_='title').text
        place_summary = item.find('span', class_='small-text text-dot-d').text.replace('\n\t\t\t\t\t\t\t\t\t\t\t', '')
        img_element = item.find('div',class_='thumb').find('img')['src']
        place_img = 'https://korean.visitseoul.net/'+img_element
        try:
            place_rating = item.find('span',class_='trip-ico').find('img')['alt']
            place_review = item.find('span',class_='trip-text').text.replace('reviews', '')
        except:
            place_rating = ""
            place_review= ""


        if place_rating == "":
            try:
                place_rating = soup_detail.find('span',class_='trip-ico').find('img')['alt']
            except:
                place_rating = ""

        if place_review == "":
            try:
                place_review = item.find('span',class_='small').text.replace('Based on'|'Traveler Reviews', '')
            except:
                place_review = ""

        #주소 가져오기
        address_dt = soup_detail.find('dt', text='주소')
        address_dd = address_dt.find_next('dd').text.strip()
        address_dd = address_dd[7:]

        print('장소명 :' + place_name)
        print('요약 :' + place_summary)
        print('이미지주소 :' + place_img)
        print('평점 :' + place_rating)
        print('리뷰수 :' + place_review)
        print('주소 :' + address_dd)
        
        
        time.sleep(3)


        #새탭 닫고 원래탭 이동
        driver.close()
        driver.switch_to.window(driver.window_handles[0])

        
        #데이터 저장
        place_list.append([place_name,place_summary,place_img,place_rating,place_review,address_dd])

        
#print(place_list)
# 파일로 저장하기
df = pd.DataFrame(place_list, columns = ['장소명','요약','장소사진','평점','리뷰수','주소'])
df.to_csv('서울관광명소_원본용.csv', encoding='utf-8-sig')
pd.read_csv
```
